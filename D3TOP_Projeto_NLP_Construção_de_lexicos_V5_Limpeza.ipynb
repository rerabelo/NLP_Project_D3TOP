```python
import re

def remover_numeros(texto):
    numeros_removidos = re.findall(r'\d+', texto)
    texto_sem_numeros = re.sub(r'\d+', '', texto)
    return texto_sem_numeros, numeros_removidos

# Remover números da coluna 'no_commas_Description_02' e obter os números removidos
df3['no_commas_Description_02'], numeros_removidos = zip(*df3['no_commas_Description_02'].apply(remover_numeros))

# Imprimir os números removidos
print("Números removidos:")
print(numeros_removidos)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def remover_palavras_raras(df, coluna_descricao, limite_frequencia):
    # Cria uma instância do CountVectorizer
    count_vectorizer = CountVectorizer()

    # Ajusta o CountVectorizer aos dados de descrição
    count_vectorizer.fit(df[coluna_descricao])

    # Obtém as palavras e suas frequências
    word_frequencies = count_vectorizer.vocabulary_

    # Filtra as palavras raras
    palavras_raras = [palavra for palavra, frequencia in word_frequencies.items() if frequencia <= limite_frequencia]

    # Remove as palavras raras da coluna de descrição
    df[coluna_descricao] = df[coluna_descricao].apply(lambda texto: ' '.join([palavra for palavra in texto.split() if palavra not in palavras_raras]))

    return palavras_raras

# Defina o limite de frequência mínima das palavras raras
limite_frequencia = 20

# Remova as palavras raras da coluna 'no_commas_Description_02'
palavras_raras_removidas = remover_palavras_raras(df3, 'no_commas_Description_01', limite_frequencia)

# Imprima as palavras raras removidas
print("Palavras raras removidas:")
print(palavras_raras_removidas)
```
