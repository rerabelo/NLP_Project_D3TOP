## **Descrição do problema**

Este projeto tem como objetivo utilizar NLP (*Natural Language Processing*) para a construção de léxicos descritivos para produtos alimentícios, e bebidas. Esse tipo de léxico fornece um vocabulário padronizado para descrever os atributos sensoriais de um produto, e por isso, é uma poderosa ferramenta para construção de análises sensoriais e pesquisas com consumidores mais assertivas. Apesar disso, a construção um léxico cara e complexa, e por isso, muitas empresas não conseguem adotar essa ferramenta para a melhoria contínua de seus produtos e formulações.


Um aspecto importante diretamente relacionado à construção de léxicos descritivos é que um léxico construído para avaliar uma categoria específica de produto em um determinado país não pode ser simplesmente traduzido e aplicado à população de outro país. Isso porque a percepção sensorial de um produto é influenciada por fatores culturais, como hábitos alimentares, preferências alimentares e métodos culinários, que podem variar muito entre países e regiões. Por exemplo, um léxico usado para avaliar queijo na França pode incluir termos específicos para variedades de queijo francês, como "cabra" ou "terroso", que podem não ser relevantes ou aplicáveis a queijo em outro país. Além disso, fatores ambientais como clima, solo e práticas agrícolas podem afetar as propriedades sensoriais de um produto, levando a diferenças na percepção sensorial entre as regiões. Por exemplo, o mesmo grão de café pode ter diferentes perfis de sabor dependendo do local onde é cultivado.


Com base nisso, a NLP pode ser uma ferramenta útil para a construção de léxicos descritivos na indústria de alimentos a partir de dados coletados diretamente dos consumidores. As indústrias poderiam, assim, ter à sua disposição léxicos eficazes para identificar características sensoriais específicas que são preferidas pelos consumidores. Do ponto de vista da indústria de alimentos e bebidas, as principais vantagens que os léxicos oferecem são listados abaixo:
*   Linguagem padronizada para descrever os atributos sensoriais como aparência, aroma, sabor, textura e sabor residual;
*   Comunicação eficaz entre as partes interessadas, que podem incluir cientistas de alimentos, analistas sensoriais, desenvolvedores de produtos, equipes de marketing, equipes de controle de qualidade, e consumidores,
*   Garantia de coleta de dados sensoriais confiáveis e consistentes.

As principais contribuições que esse projeto poderia dar para a área de ciência e tecnologia de alimentos e bebidas seriam:
*   Uma ferramenta de anotação interativa para facilitar a identificação de descritores e não descritores, que permite a criação de um conjunto de treinamento para uma arquitetura de aprendizado profundo;
*   Um pipeline para preparação de dados, anotação, treinamento e teste.

### **Normalização de texto**

Conjunto de tarefas computacionais de pré-processamento de texto que visa uniformizar os textos para uma forma mais conveniente de serem processados. A normalização de texto em geral envolve quatro tarefas: **Tokenização**, **Lematização** e **Radicalização**

#### **Capitalização de texto**

A capitalização de texto é a prática de ajustar a caixa das letras em texto. No processamento de texto e no processamento de linguagem natural (NLP), a capitalização pode às vezes carregar informações significativas, mas na maioria dos casos, o texto é convertido para letras minúsculas (como será feito a seguir) para manter a consistência e evitar duplicações baseadas em diferenças de caixa. Por exemplo, as palavras "Olá" e "olá" seriam tratadas como palavras diferentes por um computador, mesmo que sejam a mesma para um leitor humano.

```python
# Function to convert all tokens to lowercase
def lowercase_words(text):
    return [word.lower() for word in text]

# Apply function to the tokenized_Description column
df3["tokenized_Description"] = df3["tokenized_Description"].apply(lowercase_words)

print(df3)
```

### **Lematização**

A lematização é um processo na área de Processamento de Linguagem Natural (NLP) que visa reduzir uma palavra à sua forma base ou raiz, também conhecida como "lemma".

Diferentemente do processo de "stemming" (ou radicalização), que simplesmente corta as terminações das palavras esperando chegar à raiz da palavra, a lematização considera o contexto linguístico e parte do discurso da palavra para transformá-la em sua forma base.

Por exemplo, a palavra "correndo" seria reduzida para "correr" durante a lematização. Da mesma forma, "melhores" seria lematizado para "bom".

A lematização é especialmente útil no processamento de linguagem natural quando queremos analisar o significado de um texto. Ao reduzir as palavras à sua forma raiz, podemos considerar palavras diferentes que derivam do mesmo lemma como a mesma, ajudando a reduzir a complexidade do texto e a tornar a análise mais precisa.

```python
# Importar o WordNetLemmatizer do nltk
from nltk.stem import WordNetLemmatizer
import nltk

# Baixar o recurso 'wordnet' do nltk, se ainda não tiver sido baixado
nltk.download('wordnet')

# Criar o lematizador
lemmatizer = WordNetLemmatizer()

# Função para lematizar cada token de uma lista de tokens
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

# Aplicar a função na coluna "tokenized_Description"
df3["lemmatized_Description"] = df3["tokenized_Description"].apply(lemmatize_tokens)

print(df3)
```

```python
df3
```

Diante do resultado obtido observamos que o lematizador WordNetLemmatizer no NLTK utiliza o WordNet Lexical database para lematização, e que essa base de dados faz um bom trabalho de lematização para substantivos. No entanto, não é muito eficaz com adjetivos e verbos, e essa é provavelmente a razão pela qual a palavra "magnificently" não foi corretamente lematizada (Vide a primeira linha do nosso banco de dados)

Um problema comum é que o lematizador WordNet em NLTK trata tudo como substantivo a menos que especificado. Vamos adicionar informação sobre a parte de fala (POS tag) para melhorar a lematização. No entanto, o NLTK usa um conjunto diferente de tags POS, então precisamos criar uma função para mapear essas tags para as tags POS aceitas pelo WordNet (n para substantivo, v para verbo, a para adjetivo).

Vamos ajustar o código para tentar resolver esse problema:

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize, sent_tokenize

# Certifique-se de baixar os pacotes 'wordnet' e 'averaged_perceptron_tagger', se ainda não o fez
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Função para converter as tags POS do NLTK para o formato do Wordnet
def get_wordnet_pos(word):
    """Mapeia as tags POS do NLTK para o formato do Wordnet"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

# Crie uma instância do WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Agora aplique o lematizador (com POS tag) para cada token na coluna 'tokenized_Description'
df3['lemmatized_Description_01'] = df3['tokenized_Description'].apply(lambda x: [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in x])

print(df3)
```

Como vimos, a lematização ainda não está sendo eficiente. Uma outra alternativa para a lematização seria usar a biblioteca spaCy, que tem um lematizador que não necessita de mapeamento de tag POS como o NLTK.

```python
# Análise de similaridade de Jaccard

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Juntar todas as palavras em uma única string
words_01 = df3['no_stopwords_Description_01'].apply(' '.join).tolist()
words_02 = df3['no_stopwords_Description_02'].apply(' '.join).tolist()

# Cria o vetorizador de contagem, que transformará o texto em uma matriz de contagens de palavras
vectorizer = CountVectorizer().fit_transform(words_01 + words_02)

# Calcula a matriz de similaridade
similarity_matrix = cosine_similarity(vectorizer)

# Como a matriz é simétrica, podemos pegar apenas a parte superior direita da matriz
jaccard_similarity = similarity_matrix[:len(words_01), len(words_01):]

print(f"A similaridade de Jaccard média entre as duas nuvens de palavras é {jaccard_similarity.mean()}.")
```

O índice de Jaccard varia de 0 a 1, onde 0 indica que não há interseção entre os conjuntos, ou seja, eles não têm nada em comum, e 1 indica que os conjuntos são idênticos, ou seja, todas as palavras presentes em um conjunto estão presentes no outro.

No seu caso, uma similaridade de Jaccard de aproximadamente 0.123 sugere que há alguma sobreposição entre as palavras nas duas nuvens, mas não muita. Isso indica que as palavras usadas nas duas colunas da sua base de dados têm alguma semelhança, mas também têm muitas diferenças.

```python
import string
def remove_punctuation(input_list):
    # Faz uma lista de caracteres de pontuação
    punctuations = string.punctuation
    
    # Substitui a pontuação em cada palavra da lista, incluindo vírgulas extras
    no_punct = [word.translate(str.maketrans('', '', punctuations)).replace(",", "") for word in input_list]
    
    return no_punct

# Aplica a função para remover pontuações nas colunas 'no_stopwords_Description_01' e 'no_stopwords_Description_02'
df3['no_punct_Description_01'] = df3['no_stopwords_Description_01'].apply(remove_punctuation)
df3['no_punct_Description_02'] = df3['no_stopwords_Description_02'].apply(remove_punctuation)
```

```python
def remove_commas(input_list):
    # Substitui a vírgula em cada palavra da lista
    no_commas = [word.replace(",", "") for word in input_list]
    return no_commas

# Aplica a função para remover vírgulas nas colunas 'no_punct_Description_01' e 'no_punct_Description_02'
df3['no_commas_Description_01'] = df3['no_punct_Description_01'].apply(remove_commas)
df3['no_commas_Description_02'] = df3['no_punct_Description_02'].apply(remove_commas)
```
